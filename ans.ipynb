##### The following code is written according to the selenium version 3.141.0

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
import pandas as pd
import numpy as np
import json
```

```python
# Reading the file
amazon= pd.read_csv("Amazon Scraping - Sheet1.csv")
amazon.head(2)
```

```python
# Getting the driver ready
PATH = "D:\selenium\chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
# Making lists to be later added to the dictionary
iteration_found = []
titles = []
image_urls = []
prices = []
details = []
```

```python
# URl according to the assignment
url ="https://www.amazon.{country}/dp/{asin}"
```

##### The following 4 blocks of functions will scrape titles, image urls, price, details and append them to the respective lists.

```python
# Block 1
def get_titles(driver,titles):
    titles.append(driver.find_element_by_css_selector("span[id='productTitle']").text)
```

```python
# Block 2
def get_imageurls(driver,image_urls):
    try:
        image_urls.append(driver.find_element_by_id("imgBlkFront").get_attribute("src"))
    except:
        image_urls.append(driver.find_elements_by_css_selector("div.imgTagWrapper img")[0].get_attribute("src"))
```

```python
# Block 3
def get_prices(driver,prices):
    try:
        a = driver.find_element_by_css_selector('.a-color-price').text
        if '€' in a:
            prices.append(a)
            return
    except:
        pass

    try:
        b = driver.find_element_by_css_selector("span[class='a-price aok-align-center reinventPricePriceToPayMargin priceToPay']").text
        if '€' in b:
            prices.append(b)
            return
    except:
        pass

    try:
        c = driver.find_element_by_id("a-autoid-1-announce").find_element_by_class_name('a-color-base').text
        if '€' in c:
            prices.append(c)
            return
    except:
        pass

    try:
        d = driver.find_element_by_id("a-autoid-2-announce").find_element_by_class_name('a-color-base').text
        if '€' in d:
            prices.append(d)
            return
    except:
        pass

    try:
        e = driver.find_element_by_id('a-autoid-4').find_element_by_class_name('a-color-price').text
        
        if '€' in e:
            prices.append(e)
            return
    except:
        pass
    prices.append('No specified price.')
```

```python
# Block 4
def get_details(driver,details):
    
    try:
        if(driver.find_element_by_css_selector("div[id='poToggleButton']").text=='See more'):
            driver.find_element_by_css_selector("i[class='a-icon a-icon-extender-expand']").click()
    except:
        pass



    try:
        details.append(driver.find_element_by_id("detailBullets_feature_div").text)
    except:
        try:
            details.append(driver.find_element_by_css_selector("div[class='a-section a-spacing-small a-spacing-top-small']").text)
        except:
            details.append('No Details')
```

##### The following block will run a for loop over all the URLs and run the above 4 functions on all of them.

```python
#  This is a list of accept cookies in four different languages
a_c = ['accept cookies' , 'accepter les cookies','accetta i cookie', 'cookies akzeptieren','aceptar cookies']

for i in range(len(amazon)):
    driver.get(url.format(country = amazon["country"][i], asin = amazon["Asin"][i]))

    # This part will check if the url is found or not by just checking if the page has a product title or not
    try:
        if driver.find_element_by_css_selector("span[id='productTitle']").text==True:
            pass
    except:
        print(driver.current_url,'not available')
        continue

    # Append found url number into this list
    iteration_found.append(i)

    # This is to accept cookies if there is a button for that. It checks if there is an accept cookies button by checking 
    # the text of the same in the a_c list and then clicks it if it is found.
    try:
        if driver.find_element_by_css_selector("span[class='a-button-text']").text.lower() in a_c:
            driver.find_element_by_css_selector("span[class='a-button-inner']").click()
    except:
        pass

    # To fnd the title of product
    get_titles(driver,titles)

    # To find image url of product
    get_imageurls(driver,image_urls)
    
    # To get the price of the product
    get_prices(driver,prices)
    
    # To get the details. Also to expand the description by clicling on see more, if there is a see more option.
    get_details(driver,details)
```

```python
print(len(titles))
print(len(image_urls))
print(len(details))
print(len(iteration_found))
print(len(prices))
```

##### The following code makes dictionary of all 4 details.

```python
d = {}
d["titles"] = titles
d["image_urls"] = image_urls
d["details"] = details
d["prices"] = prices
```

```python
with open('output.json', 'w') as f:
    json.dump(d, f)
```
